```python
review = "⭐⭐⭐⭐⭐ GREAT PRODUCT!!! Works 100% as expected!"

def clean_review(r):
    clean = ""
    for re in r:
        if re.isalpha() or re == " ":   #.isalpha() → Only letters allowed
            clean += re

    clean = " ".join(clean.split())   # duplicate spaces ko clean
    return clean.lower()

clean_review(review)
```
