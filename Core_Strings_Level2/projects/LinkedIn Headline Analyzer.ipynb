```python
headline = "   python developer |    ai enthusiast | learning daily  "

def analyze_headline(h):
    h = h.strip()  #space 
    h = " ".join(h.split())   #extra space
    h = h.title()            # poper case

    keywords = ["Python", "Ai", "Ml", "Data", "Developer"]
    score = 0

    for word in keywords:
        if word.lower() in h.lower():
            score += 20

    total_words = len(h.split())

    return h,total_words,score

clean, words, score = analyze_headline(headline)

print("Clean Headline:", clean)
print("Total Words:", words)
print("Score:", score)
```
