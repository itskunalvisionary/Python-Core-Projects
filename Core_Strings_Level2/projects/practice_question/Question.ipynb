```python
# Q1. Split a sentence into words

s = "Python is Number One for AI"
words = s.split()
print(words)
```

```python
# Q2. Join a list of words with "-"

lst = ['Python', 'is', 'Number', 'One', 'for', 'AI']
result = "-".join(lst)
print(result)
```

```python
# Q3. Replace “AI” with “Artificial Intelligence”

a = "AI will change the world"
print(a.replace("AI","Artificial Intelligence"))
```

```python
# Q4. Detect substring location

s = "Hello Python World"
print(s.find("Python"))
```

```python
# Q5. Count occurrences of “a” in a string

a = "kunal change ai"
print(a.count("a"))
```

```python
# Q6. Use f-strings to format output

name = input("enter your name: ").title().strip()
age = int(input("enter you age: "))
print(f"My name is {name} and I am {age} years old.")
```

```python
# Q7. Check if review starts with “good”

review =  "good product"
print(review.startswith("good"))
```

```python
# Q8. Remove special characters from filename

file_name = "my*file@#.txt"

clean = ""
for f in file_name:
    if f.isalnum() or f in "._":   #isalnum() = alphabet or number.
        clean += f

print(clean)
```

```python
# Q9. Convert string to camelCase

s = "hello world program".split()
camel = s[0] + "".join(word.capitalize() for word in s[1:])
print(camel)
```

```python
# Q10. Predict Output

print("python".replace("p", "P").upper())

# "python".replace("p", "P") → "Python"
# "Python".upper() → "PYTHON"
```
