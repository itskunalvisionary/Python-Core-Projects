```python
#Input Section

original_price = float(input("enter product price: "))
discount_percent = float(input("enter discount percent: "))
gst_percent = float(input("enter gst percent: "))
delivery_charge = float(input("enter delivery charge: "))

#Calculation Section

discounted = original_price - (original_price*discount_percent/100)
gst = discounted + (discounted * gst_percent/100)
final_price = gst_percent + delivery_charge + discounted

print(f"Original Price: {original_price:.2f}")
print(f"Discounted: {discounted}")
print(f"GST Added: {gst:.2f}")
print(f"Final Price: {final_price}")
```
