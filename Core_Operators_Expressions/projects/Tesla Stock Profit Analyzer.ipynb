```python
# Tesla Stock Profit Analyzer

# Inputs
buy_price = float(input("Enter buy price per share: "))
sell_price = float(input("Enter sell price per share: "))
shares = int(input("Enter number of shares held: "))

# Calculations
total_buy = buy_price * shares
total_sell = sell_price * shares
profit = total_sell - total_buy

# Status classification
if profit > 0:
    status = "GAIN"
elif profit < 0:
    status = "LOSS"
else:
    status = "BREAK-EVEN"

# Output
print("--- Trade Summary ---")
print(f"Total Buy:   {total_buy}")
print(f"Total Sell:  {total_sell}")
print(f"Profit:      {profit}")
print(f"Status:      {status}")
```
