```python
# PayPal Fraud Detection Engine

# Sample transaction data
amount = int(input("enter the amount of payment: "))
time_between_txn = 8 # in seconds

is_foreign_currency = True
is_odd_amount = amount % 2 != 0

# Rule-based risk classification

if amount > 50000:
    risk_level = "HIGH"
    reason = "amount exceeds threshold"

elif time_between_txn > 10:
    risk_level = "SUSPICIOUS"
    reason = "multiple transactions in < 10 sec"

elif is_odd_amount and is_foreign_currency:
    risk_level = "REVIEW"
    reason = "odd amount + foreign currency"

else:
    risk_level = "LOW"
    reason = "no flag triggered"

# output 
print("--- Transaction Risk Report ---")
print(f"Transaction Risk Level: {risk_level}")
print(f"Reason: {reason}")
```
