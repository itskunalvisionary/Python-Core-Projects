```python
# Crypto Reward Simulator

# Inputs
principal = float(input("Enter principal amount: "))
apr = float(input("Enter APR (e.g., 12 for 12%): ")) / 100
duration = float(input("Enter duration in years: "))
n = int(input("Enter compounding frequency per year (e.g., 12 for monthly): "))

# Compound interest calculation
final_amount = principal * (1 + apr / n) ** (n * duration)
reward = final_amount - principal

# Output
print("\n--- Staking Summary ---")
print(f"Principal:       {principal}")
print(f"Final Amount:    {round(final_amount, 2)}")
print(f"Total Reward:    {round(reward, 2)}")
```
