```python
# Find the discount %: original vs selling price.

original_price = 1000
selling_price = 750

find_discount = (original_price - selling_price) / original_price*100

print("Discound Card")
print("_"*50)
print(f"Original Price: ₹{original_price}")
print(f"Selling Price: ₹{selling_price}")
print("-"*50)
print(f"Discound: {find_discount}%")
```

```python
# Calculate compound interest using **.

# Inputs
P = 1000       # Principal
r = 0.05       # Annual interest rate (5%)
n = 4          # Compounded quarterly
t = 3          # Time in years

# Compound interest calculation
A = P * (1 + r / n) ** (n * t)

print("Final amount:", round(A, 2))
```

```python
#  Bitwise operations on: a = 12, b = 5

a = 12
b = 5

print(a & b)  # AND
print(a | b)  # OR
print(a ^ b)  # XOR
print(a << 1) # Left shift
print(b >> 1) # Right shift
```

```python
# Check if number is odd without using %

number = 13

if number & 1:
    print("Odd")

if number % 2 != 0:
    print("Odd")
```
