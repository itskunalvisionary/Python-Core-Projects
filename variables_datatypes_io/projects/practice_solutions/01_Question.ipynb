```python
# Calculate simple interest (SI = P × R × T / 100).

P = float(input("Enter the principal amount (P): "))
R = float(input("Enter the rate of interest (R): "))
T = float(input("Enter the time period in years (T): "))

# Calculate simple interest
SI = (P * R * T) / 100

# Display the result using an f-string
print(f"The Simple Interest is: {SI}")
```

```python
# Convert seconds → minutes and seconds.

total_seconds = int(input('enter second: '))

#Convert
minutes = total_seconds // 60
seconds = total_seconds % 60

#result
print(f"{total_seconds} seconds is {minutes} minute(s) and {seconds} second(s).")
```

```python
# Area of a circle from user input (radius).

import math
user_radius = float(input("enter the radius: "))

area = math.pi * user_radius **2

print(f"The Area of Circle is {area:.2f}")
```
