```python
def extract_username(url):
    """
    Extracts the Instagram username from a profile URL.
    Example:
    url = 'https://www.instagram.com/elonmusk/'
    output -> 'elonmusk'
    """

    # Remove trailing slash if it exists
    if url.endswith("/"):
        url = url[:-1]

    # Username is the last segment after '/'
    return url.split("/")[-1]
```

```python
# ---------- Testing / Demo ----------
urls = [
    "https://instagram.com/elonmusk",
    "https://www.instagram.com/cristiano/",
    "instagram.com/taylorswift"
]

for u in urls:
    print(extract_username(u))
```
