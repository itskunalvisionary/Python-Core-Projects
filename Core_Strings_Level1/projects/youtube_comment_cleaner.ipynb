```python
def clean_comment(comment):
    """
    Cleans messy YouTube comments by:
    - Removing emojis (keep normal characters only)
    - Removing digits
    - Removing special characters
    - Removing extra spaces
    """

    clean = ""
    for ch in comment:
        # Keep alphabets and spaces only
        if ch.isalpha() or ch == " ":
            clean += ch

    # Remove extra spaces
    clean = " ".join(clean.split())

    return clean.title()
```
