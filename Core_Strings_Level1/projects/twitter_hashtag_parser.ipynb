```python
def extract_hastags(text):
    """
    Extracts hashtags from a Twitter-style text.
    Example:
    '#AI #MachineLearning #Python' -> ['AI', 'MachineLearning', 'Python']
    """

    words = text.split()
    hastags = []

    for w in words:
        if w.startswith("#") and len(w) > 1:
            hastags.append(w[1:])  # Remove '#'

    return hastags
```

```python
# ---------- Testing / Demo ----------
tweet = """Toady World is Change fast Because Nothing 
        #AI #MachineLearning #OpenAI #Python #2025"""

print(extract_hastags(tweet))
```
