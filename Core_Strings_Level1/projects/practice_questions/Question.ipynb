```python
# 4Ô∏è‚É£ Count vowels in a sentence

sentence = "Machine Learning"

count = 0
for ch in sentence.lower():
    if ch in "aeiou":
        count += 1

print("---Count Vowel in Sentence---")
print("Sentence:",sentence)
print("Count:",count)
```

```python
# 6Ô∏è‚É£ Check if input is palindrome

a = "cricket"
print(a == a[::-1])

print("-"*50)
b = "level"
print(b == b[::-1])
```

```python
# 7Ô∏è‚É£ Merge two strings alternatively


a = "ABC"
b = "123"
result = ""
for i in range(max(len(a), len(b))):
    if i < len(a): result += a[i]
    if i < len(b): result += b[i]
print(result)
```

```python
# 8Ô∏è‚É£ Remove digits from a string

s = "python3xpert"

result = ""

for ch in s:
    if not ch.isdigit():
        result += ch


print(result)
```
