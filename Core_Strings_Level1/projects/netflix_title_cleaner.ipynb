```python
def clean_movie_title(title):
    """
    Cleans messy Netflix movie titles.
    Steps:
    - Strip side spaces
    - Remove extra internal spaces
    - Remove year in parentheses
    - Convert to title case
    """

    # Step 1: Remove side spaces
    title = title.strip()

    
    # Step 2: Remove content inside parentheses (year)
    if "(" in title:
        title = title.split("(")[0].strip()

    # Step 3: Remove extra internal spaces
    title = " ".join(title.split())

    # Step 4: Proper capitalization
    return title.title()
```

```python
# ---------- Testing / Demo ----------
movies = [
    "inception (2010)",
    "the dark knight   ",
    "  MONEY HEIST - SEASON 1",
    "breaking bad (2008)",
]

for m in movies:
    print(clean_movie_title(m))
```
