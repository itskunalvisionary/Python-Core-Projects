```python
def neflix_age_checker(age,rating):
    if rating == "U":
        return "Allowed"
    
    if rating == "U/A" and age >= 12:
        return "Allowed"
    
    if rating == "A" and age >= 13:
        return "Allowed"
    
    if rating == "18+" and age >= 18:
        return "Allowed"
    
    return "Blocked"

print(neflix_age_checker(15, "U/A"))  #Output: Allowed
print(neflix_age_checker(10, "A"))    #Output: Blocked
    
    
```
