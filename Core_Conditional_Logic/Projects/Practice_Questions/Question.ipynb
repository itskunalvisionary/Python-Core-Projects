```python
# Q1. Grade Calculator (Based on Marks)

try:
    marks = float(input("enter you marks to see grade: "))

    if marks >= 90:
        grade = "A+"
    elif marks >= 75:
        grade = "A"
    elif marks >= 60:
        grade = "B"
    elif marks >= 45:
        grade = "C"
    else:
        grade = "Fail"

    print("Grade:", grade)
except ValueError as ve:
    print({ve})
```

```python
# Q4. Password Strength Checker

password = "Abcdefgh12345"

if len(password) <= 6:
    print("Weak") 
elif password.isalpha() or password.isdigit():
    print("Moderate")
else:
    print("Strong")
```

```python
# Q5. Leap Year Check

year = 2024

if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print("Leap Year")
else:
    print("Not Leap Year")
```

```python
# Q9. Determine Triangle Type

a,b,c = 5,5,5

if a == b == c:
    print("Equilateral")
elif a == b or b == c or a == c:
    print("Isosceles")
else:
    print("Scalene")
```
