```python
def ride_far(distance, traffic, time, surge, memeber):

    base = 50
    cost = distance * 10

    if traffic == "high":
        cost += 30
    elif traffic == "medium":
        cost += 20

    if time in ["night","peak"]:
        cost *= surge

    if memeber:
        cost -= 40

    return (base,cost)

print(ride_far(10, "medium", "peak", 1.5, True))
```
