```python
def delivery_time(distance, rain, traffic):

    time = 0

    if distance <3:
        time += 20
    elif distance <6:
        time += 30
    else:
        time += 45

    if rain:
        time +=10

    if traffic:
        time += 15

    return time

print("Delivery Time:",delivery_time(4, True, True), "mintues ")
```
